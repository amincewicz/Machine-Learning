```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split,cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import svm
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)
import pickle
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
#Provide the information necessary for the header
header = ["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country", "class"]
```

```python
#Read in data from the file and add the header
df = pd.read_csv("adult.data", header = None, names = header)
```

```python
df.head()
```

```python
#As the class includes sting value, get dummies to get information in forms of 1s and 0s
tempdf = pd.get_dummies(df["class"]) 
```

```python
#Concatenate all the date to create one dataframe
newdf = pd.concat([df,tempdf], axis = "columns")
```

```python
newdf.head()
```

```python
#Drop the two unnecessary column
newdf.drop(columns=newdf.columns[-2],axis=1,inplace=True)
newdf.drop(columns='class',axis=1,inplace=True)
```

```python
newdf.head()
```

```python
#Select the columns that have categorical data
newdf_cat = newdf.filter(["workclass", "education", "martial-status", "occupation", "relationship", "race", "sex", "native-country", "outcome"])
```

```python
#Encode the categorical data
encoder = preprocessing.LabelEncoder()
done_df = newdf.apply(encoder.fit_transform)
done_df.head()
```

```python
done_df.head()
```

```python
#Determine which data is X and which is Y
Y = done_df[done_df.columns[-1]]
X = done_df.drop(done_df.columns[-1], axis = 1)
```

```python
#Split the data and determine training and tersting sizes
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=1)
```

Logistic Regression

```python
#Select logistic regression as the first classifier and fit it
log_reg = LogisticRegression(solver='lbfgs') 
log_reg.fit(X, Y)              
log_reg.coef_               
```

```python
#Calculate and print the accurqacy score
Y_pred = log_reg.predict(X_test)
score =accuracy_score(Y_test,Y_pred)
print(score)
```

```python
#Cross validate
log_reg_cvs = cross_val_score(log_reg, X_train, Y_train, cv=10, n_jobs=-1, scoring="accuracy")
```

SVM model

```python
#Select SVM as the second classifier and fit it
model = svm.SVC(kernel='linear')
model.fit(X, Y)
```

```python
#Calculate the prediction
Y_pred = model.predict(X_test)
```

```python
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```

```python
#Cross validate
svm_cvs = cross_val_score(model, X_train, Y_train, cv=10, n_jobs=-1, scoring="accuracy")
```

Hard and Soft Voting

```python
#Determine the three classifiers to use for hard voting
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC()
```

```python
#Start with hard voting
voting_clf = VotingClassifier(
    estimators=[("lr", log_clf), ("rf", rnd_clf), ("svc", svm_clf)],
    voting="hard")
voting_clf.fit(X_train, Y_train)
```

```python
#Calculate the prediction accuracy
pred_accuracy_hard=cross_val_predict(voting_clf, X_train, Y_train, cv=10)
```

```python
#Determine the three classifiers to use for soft voting
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC(probability=True)
```

```python
voting_clf2 = VotingClassifier(
    estimators=[("lr", log_clf), ("rf", rnd_clf), ("svc", svm_clf)],
    voting="soft",flatten_transform=False)
voting_clf2.fit(X_train, Y_train)
```

```python
# Cross validate
```

```python
pred_accuracy_soft=cross_val_predict(voting_clf2, X_train, Y_train, cv=10)
```
